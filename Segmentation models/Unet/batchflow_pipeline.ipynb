```python
%load_ext autoreload
%autoreload 2
%env CUDA_VISIBLE_DEVICES = 1

from torch import nn

import batchflow
from batchflow import *
from batchflow.models.torch import *
from batchflow.models.torch.layers import *
from batchflow.opensets import PascalSegmentation
from batchflow.research.utils import plot_images
from batchflow import pylint_notebook
from batchflow.models.torch.losses import CrossEntropyLoss
```

```python
DEVICE = 'cuda:0'
NUM_CLASSES = 22
IMAGE_SHAPE = (3, 256, 256)
BATCH_SIZE = 32
N_ITERS = 30
BAR = True
```

```python
model_config = {'inputs_shapes': IMAGE_SHAPE,
                
    'order': ['encoder', 'embedding', 'decoder', 'head'], 

    'encoder': {              
        'type': 'encoder',
        'order':['blocks', 'skip', 'downsample'],
        'num_stages': 4,     
        'blocks':{'layout':'cna cna', 
                  'channels':[64, 128, 256, 512], 
                  'kernel_size':3,
                  'padding': 1, 
                  'stride': 1, 
                  'bias': False},   
        'downsample': {'layout': 'p', 
                       'kernel_size': 2,
                       'stride': 2, 
                       'channels': 'same * 2'},
    },
    'embedding': {  
        'input_type': 'list',
        'input_idx': -1,
        'output_type': 'list',
        'layout': 'cna cna',
        'kernel_size': 3,
        'padding': 1,
        'channels': 1024,
    },
    'decoder': {
        'skip': True,
        'indices': [3, 2, 1, 0],
        'type': 'decoder',
        'num_stages': 4,
        'order': ['upsample', 'combine', 'blocks'],
        'blocks': {'layout':'cna cna', 'channels':[512, 256, 128, 64], 'kernel_size': 3, 
                   'stride': 1, 'padding': 1, 'bias': False},                                      
        'upsample': {'layout': 't', 'kernel_size': 2, 'stride': 2, 'channels': 'same // 2', 'bias': True},  
        'combine': {'op': 'concat'},   'force_resize': False},             
        
    'head': {                     
        'layout': 'c', 
        'channels': NUM_CLASSES,
        'kernel_size':1,
        'stride': 1, 
        'padding': 0, 
        'bias': True},
                
    # 'loss': CrossEntropyLoss(ignore_index=255),
    'loss': 'ce',
    'device': DEVICE,
    'optimizer': {'name': 'Adam', 'lr': 1e-4}
    
}
```

```python
class PascalBatch(ImagesBatch):
    components = 'images', 'masks'
    
    def __init__(self, index, *args, **kwargs):
        super().__init__(index, *args, **kwargs)
                    
    @apply_parallel(src='images', target='for')
    def resize_min_side(self, image, size, *args, **kwargs):
        h, w = image.size
        if h > w:
            new_size = (int(h * size / w), size)
        else:
            new_size = (size, int(w * size / h))
        
        return image.resize(new_size, *args, **kwargs)
    
    @apply_parallel(src='images', target='for') 
    def normalize(self, image, mean, std):
        return (image - mean) / (std + 1e-7) 
    
    @apply_parallel(src='masks', target='for')
    def squeeze_labels(self, image):
        return image.squeeze()
    
    @apply_parallel(src='masks', target='for')
    def rename_extra_label(self, image):
        image[image == 255] = 21.
        return image
```

```python
dataset_pascal = PascalSegmentation(shuffle=True, batch_class=PascalBatch)
```

```python
batch_gen = dataset_pascal.gen_batch(1, n_iters=len(dataset_pascal))
classes = np.zeros(256, dtype=int)
for i in range(len(dataset_pascal)):
    batch = next(batch_gen)
    batch_classes, counts = np.unique(np.array(batch.masks[0], dtype=int), return_counts=True)
    classes[batch_classes] += counts
```

```python
def plot_batch_example(dataset):
    batch = dataset.p.next_batch(2)
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(17, 10))
    ax1.imshow(batch.images[1])
    ax1.axis('off')
    ax2.imshow(batch.masks[1])
    ax2.axis('off')
    
plot_batch_example(dataset_pascal);
```

```python
def get_norm_stats(dataset, batch_size=64, channels='first'):
    ppl = dataset.p.to_array(src='images', dst='images')
    images = ppl.next_batch(batch_size).images
    mean = np.zeros(3)
    std = np.zeros(3)
    for i in range(batch_size):
        mean += images[i][..., 0].mean(), images[i][..., 1].mean(), images[i][..., 2].mean()
        std  += images[i][..., 0].std(), images[i][..., 1].std(), images[i][..., 2].std()
    if channels == 'first':
        mean = mean.reshape(3, 1, 1)
        std = std.reshape(3, 1, 1)
    return  mean / batch_size, std / batch_size
```

```python
mean, std = get_norm_stats(dataset_pascal)  

preprocessing_template = (
                            Pipeline() 
                            .resize_min_side(size=256, src=['images', 'masks'], dst=['images', 'masks'])  
                            .crop(shape=(256, 256), origin='center', src=['images', 'masks'], dst=['images', 'masks'])
                            .to_array(src=['images', 'masks'], dst=['images', 'masks'], channels='first')
                            .normalize(mean=mean, std=std, src='images', dst='images')
                            .squeeze_labels(src='masks', dst='masks')
                            .rename_extra_label(src='masks', dst='masks')
                        )
```

```python
preprocessing_pipeline = preprocessing_template << dataset_pascal.train
```

```python
batch = preprocessing_pipeline.next_batch(1)
print(batch.images[0].shape, batch.masks[0].shape, batch.masks[0].dtype)
```

```python
train_template = (  
    preprocessing_template +
    Pipeline()
        .init_variable('loss_history', [])
        .init_model(name='Unet', model_class=TorchModel, config=model_config)

        .train_model('Unet',
                     inputs=B.images, targets=B.masks,
                     outputs='loss', save_to=V('loss_history', mode='a'))
)
```

```python
train_pipeline = train_template << dataset_pascal.train
```

```python
notifier = {
    'bar': 'n', 'frequency': 1,
    'graphs': ['loss_history'],
}
```

```python
train_pipeline.run(BATCH_SIZE, n_epochs=N_ITERS, notifier=notifier)
```

```python
# val pipeline

test_notifier = {
    'frequency': 1,
    'graphs': ['gpu'],
}

test_pipeline = (dataset_pascal.test.p
                .init_variable('predictions', default=[])
                .init_variable('metrics')
                .import_model('Unet', train_pipeline)
                
                .resize_min_side(size=256, src=['images', 'masks'], dst=['images', 'masks'])  
                .crop(shape=(256, 256), origin='center', src=['images', 'masks'], dst=['images', 'masks'])
                .to_array(src=['images', 'masks'], dst=['images', 'masks'], channels='first', dtype='float32')
                .squeeze_labels(src='masks', dst='masks')
                .rename_extra_label(src='masks', dst='masks')
                
                .predict_model('Unet',
                               inputs=B.images,
                               outputs='predictions',
                               save_to=V('predictions', mode='a'))
                .gather_metrics(metrics_class='segmentation',
                                fmt='proba', axis=1, 
                                targets=B.masks, predictions=V.predictions[-1],
                                save_to=V('metrics', mode='a'))
                      
                # .run(BATCH_SIZE, shuffle=False, n_epochs=1, drop_last=False, bar=BAR, notifier=test_notifier)
                )
```

```python
metrics = test_pipeline.get_variable('metrics')
```

```python
metrics.evaluate('jaccard')
```
